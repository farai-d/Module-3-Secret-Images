Cell 2 - CBC Decryption Code 2

```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import unpad
from PIL import Image
import matplotlib.pyplot as plt

key_string = "world series2026"

key = key_string.encode()

with open("cbc-aes-picture.img","rb") as f:
    iv = f.read(16)          
    ciphertext = f.read()    

cipher = AES.new(key, AES.MODE_CBC, iv)

plaintext = cipher.decrypt(ciphertext)

try:
    plaintext = unpad(plaintext,16)
except:
    pass

with open("recovered_image.jpg","wb") as f:
    f.write(plaintext)

print("Decryption complete -> recovered_image.jpg")


img = Image.open("recovered_image.jpg")

plt.imshow(img)
plt.title("Recovered Image")
plt.axis("off")
plt.show()
```
